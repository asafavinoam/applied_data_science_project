## sql to csv

The perpose of the code below is only to create csv files from sql database we work with. The reason for this transformation is that github not allow us to upload more than 100MB and the sql is ~300MB.

#### General Impotrs:

```python
import sqlite3
import numpy as np
import pandas as pd
import itertools
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read all DB's:

```python
main_db = sqlite3.connect('databases/database.sqlite')
```

```python
leagues_db = pd.read_sql("SELECT * FROM League", main_db)
leagues_db.to_csv("databases/sql_leagues_db.csv")
leagues_db.head()
```

```python
teams_db = pd.read_sql("SELECT * FROM Team", main_db)
teams_db.to_csv("databases/sql_teams_db.csv")
teams_db.head()
```

```python
all_leagues = dict(zip(list(leagues_db["id"]),list(leagues_db["name"])))
```

#### Create csv file for each league:

```python
matches_db = {}

for league_id,league_name in all_leagues.items():
    matches_db[league_name]         = pd.read_sql("SELECT * FROM Match where league_id="+str(league_id), main_db)
    matches_db[league_name]["date"] = matches_db[league_name]["date"].apply(lambda x: x.split(" ")[0])
    matches_db[league_name].to_csv("databases/sql_matches_db/"+ league_name.replace(" ","_") + ".csv")
```
